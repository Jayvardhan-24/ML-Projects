```python
for col in df.columns:
  if(df[col].dtype!='object'):
    sns.boxplot(df[col])
    plt.show()
```

```python
out_list=['age','education-num','fnlwgt','hours-per-week']
for col in out_list:
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)

  IQR=Q3-Q1

  LB=Q1-1.5*(IQR)
  UB=Q3+1.5*(IQR)

  df=df[(df[col]>=LB) & (df[col]<=UB)]
```
